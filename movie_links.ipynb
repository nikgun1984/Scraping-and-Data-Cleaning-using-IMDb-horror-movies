```python
# libraries used
import warnings
warnings.filterwarnings('ignore')
import re
from urllib.request import urlopen
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
from requests import get
```

```python
movie_name=[] # for storing movie names
movie_links=[] # for storing their IMDB links
```

```python
urls = []
pages = 1
url = 'https://www.imdb.com/search/title?title_type=feature&num_votes=1000,&genres=horror&start='
print(urls)
#response = get(url)
#soup = BeautifulSoup(response.text, 'lxml')
for i in range(20):
    urls.append(url+str(pages))
    pages+=50
    print(urls[i])

#looping through containers
for i in urls:
    response = get(i)
    soup = BeautifulSoup(response.content, 'html.parser')
    movie_containers = soup.find_all('div', class_= 'lister-item-content')
    
    # Extract data from individual movie container
    for container in movie_containers:
        
        if container.find('div', class_ = 'ratings-imdb-rating') is not None:


            name = container.h3.a.text
            #print(name)
            movie_name.append(name)
            movies = container.select(".lister-item-header a")
            movie_l = ["http://imdb.com"+i.attrs["href"] for i in movies]
            movie_links.extend(movie_l)

test_df = pd.DataFrame(data = {
    'name' : movie_name,
    'link' : movie_links
})

test_df.to_csv("links.csv", sep=",", encoding="utf-8")
test_df
```
