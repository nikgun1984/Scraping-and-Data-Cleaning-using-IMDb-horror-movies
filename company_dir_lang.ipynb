```python
# libraries used
import warnings
warnings.filterwarnings('ignore')
import re
from urllib.request import urlopen
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
from requests import get
```

```python
movie_name=[] # for storing movie names
movie_links=[] # for storing their IMDB links
        
```

```python
urls = []
pages = 1
url = 'https://www.imdb.com/search/title?title_type=feature&num_votes=1000,&genres=horror&start='
print(urls)
#response = get(url)
#soup = BeautifulSoup(response.text, 'lxml')
for i in range(20):
    urls.append(url+str(pages))
    pages+=50
    print(urls[i])

#looping through containers
for i in urls:
    response = get(i)
    soup = BeautifulSoup(response.content, 'html.parser')
    movie_containers = soup.find_all('div', class_= 'lister-item-content')
    
    # Extract data from individual movie container
    for container in movie_containers:
        
        if container.find('div', class_ = 'ratings-imdb-rating') is not None:


            name = container.h3.a.text
            #print(name)
            movie_name.append(name)
            movies = container.select(".lister-item-header a")
            movie_l = ["http://imdb.com"+i.attrs["href"] for i in movies]
            movie_links.extend(movie_l)
            
            #all_links = soup.find("h3", class_="'lister-item-header")


            #print(type(movies[0]))
            #print(movies[0])
            #print(container.get('href'))
            #movie = [title.text for title in movies]
            #movie_name.extend(movie)
            #movie_l = ["http://imdb.com"+title["href"] for title in movies if title != None]
            
            #movie_links.extend(movie_l)

#print(len(movie_links))

print(movie_links)
print(any(x is None for x in movie_links))

dirct=[]     # Diector
country=[]   # Country
lang=[]
company1 = []
company2 = []
company3 = []


def getlinks(links):
    bsobj0 = BeautifulSoup(urlopen(links))
    country.append(bsobj0.find('div', {'class':'article', 'id':'titleDetails'}).find_all('div')[1].find('a').text)
    #try:
    #    bbfc.append(bsobj0.find('meta', {'itemprop':'contentRating'}).attrs['content']) #bbfc rating
    #except AttributeError:
    #    bbfc.append(np.nan)
    dirct.append(bsobj0.find_all('div', {'class':'credit_summary_item'})[0].a.text) #director
    if bsobj0.find("h4", text="Production Co:") is not None:
        all_links = bsobj0.find("h4", text="Production Co:").parent
        all_links = all_links.text.split("\n")[2].split(',')
        list_companies = [x.strip(' ') for x in all_links]
        company1.append(list_companies[0])
        if len(list_companies)>1:
            company2.append(list_companies[1])
            if len(list_companies)>2:
                company3.append(list_companies[2])
            else:
                company3.append(np.nan)
        else:
            company2.append(np.nan)
            company3.append(np.nan)
    else:
        company1.append(np.nan)
        company2.append(np.nan)
        company3.append(np.nan)

for i in movie_links:
    getlinks(i)

print(len(dirct),len(country))
test_df = pd.DataFrame(data = {
    'name' : movie_name,
    'country' : country,
    'Director' : dirct,
    'Company1' : company1,
    'Company2' : company2,
    'Company3' : company3
})

test_df['country'].fillna("Null", inplace = True)
test_df['Director'].fillna("Null", inplace = True)
test_df['Company1'].fillna("Null", inplace = True) 
test_df['Company2'].fillna("Null", inplace = True)
test_df['Company3'].fillna("Null", inplace = True)


test_df.to_csv("other_info.csv", sep=",", encoding="utf-8")
test_df
```

```python
from BeautifulSoup import BeautifulSoup
import urllib2
import re

def getLinks(url):
    html_page = urllib2.urlopen(url)
    soup = BeautifulSoup(html_page)
    links = []

    for link in soup.findAll('a', attrs={'href': re.compile("^http://")}):
        links.append(link.get('href'))

    return links

print( getLinks("https://arstechnica.com") )
```
